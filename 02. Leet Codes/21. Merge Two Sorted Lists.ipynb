### <span style="background-color: yellow;">21. Merge Two Sorted Lists
</span>

* **Difficulty** - Easy 

* **Description** - You are given the heads of two sorted linked lists list1 and list2. Merge the two lists in a one sorted list. The list should be made by splicing together the nodes of the first two lists. Return the head of the merged linked list. 


* **Examples -**  
1. Input: list1 = [1,2,4], list2 = [1,3,4]  
Output: [1,1,2,3,4,4]
2. Input: list1 = [], list2 = []  
Output: []
3. Input: list1 = [], list2 = [0]  
Output: [0]


* **Constraints -**
- The number of nodes in both lists is in the range [0, 50].
- -100 <= Node.val <= 100
- Both list1 and list2 are sorted in non-decreasing order.


![Merge Ex 1](https://assets.leetcode.com/uploads/2020/10/03/merge_ex1.jpg)


```python
class ListNode:                              # Refer the below code cell for linked lists "l1" and "l2"                 
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
                                                                                 # Tracing: 
class Solution:                                                                  
    def mergeTwoLists(self, l1: ListNode, l2: ListNode) -> ListNode:             # mergeTwoLists(l1, l2)
        # Create a dummy node as the head of the merged list
        dummy = ListNode(0)                                                      # dummy = ListNode(0)
        curr = dummy                                                             # curr = dummy    i.e. curr = ListNode(0)

        # Compare the values of the two lists and merge them in sorted order
        while l1 and l2:                                                         # while l1 and l2 :                 # while l1.next and l2:             # while l1.next and l2.next              # while l1.next.next and l2.next         # while l1.next.next and l2.next.next         # while l1.next.next.next and l2.next.next (False)
            if l1.val <= l2.val:                                                 # if 1 <= 1                         # if 2 <= 1 (False)                 # if 2 <= 3                              # if 4 <= 3 (False)                      # if 4 <=4                                    
                curr.next = l1                                                   # curr.next = l1                                                        # curr.next = l1.next                                                             # curr.next = l1.next.next                    
                l1 = l1.next                                                     # l1 = l1.next i.e. ListNode(2)                                         # l1 = l1.next.next i.e. ListNode(4)                                              # l1 = l1.next.next.next (does not exist)     
            else:                                                                                                    # else:                                                                      # else:
                curr.next = l2                                                                                       # curr.next = l2                                                             # curr.next = l2.next                    
                l2 = l2.next                                                                                         # l2 = l2.next i.e. ListNode(3)                                              # l2 = l2.next.next i.e. ListNode(4)       
            curr = curr.next                                                     # curr = l1 = ListNode(1)           # curr = l2 = ListNode(1)           # curr = l1.next = ListNod(2)            # curr = 12.next = ListNode(3)           # curr = l1.next.next = ListNode(4)           # The whole while loop is skipped    

        # Append any remaining nodes from either list
        if l1:                                                                                                                                                                                                                                                                           # if l1: (False : Since we know l1 =l1.next.next.next does not exist)                                   
            curr.next = l1
        if l2:                                                                                                                                                                                                                                                                           # if l2: (l2 = l2.next.next)                                  
            curr.next = l2                                                                                                                                                                                                                                                               # curr.next = l2.next.next = ListNode(4)                      

        # Return the merged list starting from the next of the dummy node
        return dummy.next                                                                                                                                                                                                                                                                                                                                          # return dummy.next = ListNode(1)
```

#### <span style="background-color: yellow;">Explaining the above code -</span>

The provided code defines a **`ListNode`** class and a **`Solution`** class with a method **`mergeTwoLists`**. This method merges two linked lists **`l1`** and **`l2`** into a single linked list in sorted order.

Let's go through the code step by step and trace its execution:

1. The **`ListNode`** class is defined to represent a node in a linked list. Each node has a value (**`val`**) and a pointer to the next node (**`next`**).

2. The **`Solution`** class is defined with a method **`mergeTwoLists`** that takes two linked lists **`l1`** and **`l2`** as input and returns the merged linked list.

3. Inside the **`mergeTwoLists`** method:
   - A dummy node is created as the head of the merged list, initialized with a value of 0. The **`dummy`** node helps simplify the merging process.
   - A variable **`curr`** is initialized to point to the dummy node. This **`curr`** pointer will be used to traverse the merged list and add nodes.
   
4. A while loop is started to compare the values of nodes from **`l1`** and **`l2`** and merge them in sorted order. The loop continues as long as both **`l1`** and **`l2`** have remaining nodes.
   - If the value of the current node in **`l1`** is less than or equal to the value of the current node in **`l2`**, the **`curr`** node's next pointer is set to the current node in **`l1`**, and **`l1`** moves to its next node.
   - Otherwise, if the value of the current node in **`l2`** is smaller, the **`curr`** node's next pointer is set to the current node in **`l2`**, and **`l2`** moves to its next node.
   - In both cases, the **`curr`** pointer moves to the next node in the merged list.

5. After the while loop ends, there may be remaining nodes in either **`l1`** or **`l2`**. The code appends these remaining nodes to the merged list by updating the **`curr`** pointer accordingly.

6. Finally, the merged list is returned, starting from the next node of the dummy node.

This code efficiently merges two sorted linked lists into a single sorted linked list.

```python
# Example 1 - 

# Creating and printing linked lists 

# -----Creating Lists---------

# Creating sample linked list l1: 1 -> 2 -> 4
l1 = ListNode(1)
l1.next = ListNode(2)
l1.next.next = ListNode(4)

# Creating sample linked list l2: 1 -> 3 -> 4
l2 = ListNode(1)
l2.next = ListNode(3)
l2.next.next = ListNode(4)

# -----Printing Lists---------

# Printing the elements of l1                                         # Tracing: 
current = l1                                                          # current = l1 = 1 
print("L1 Linked List:")                                              # print: L1 Linked List:  (o/p)
while current and current.next:                                       # while l1 and l1.next:  - i.e., while 1 and 2:     # while 2 and 4:                                                                # while 4 and "" (False)
    print(current.val, end=" -> ")                                    # print: 1 ->                                       # print: 1 -> 2 -> (Here 1 -> is already printed in the previous iteration)     # we come out of the loop
    current = current.next                                            # current = l1.next = 2                             # current = l1.next.next
print(current.val)  # Print the last element of the merged list                                                                                                                                           # print: 1 -> 2 -> 4 (we get l1 linked list)

print("")                                                             # Printing an empty line                          

# Printing the elements of l2
current = l2                                                          # current = l2 = 1                                   
print("L2 Linked List:")                                              # print: L2 Linked List:  (o/p)                     
while current and current.next:                                       # while l2 and l2.next:  - i.e., while 1 and 3:     # while 3 and 4:                                                                # while 4 and "" (False)
    print(current.val, end=" -> ")                                    # print: 1 ->                                       # print: 1 -> 3 -> (Here 1 -> is already printed in the previous iteration)     # we come out of the loop
    current = current.next                                            # current = l2.next = 3                             # current = l2.next.next
print(current.val)  # Print the last element of the merged list                                                                                                                                           # print: 1 -> 3 -> 4 (we get l2 linked list)

print("")                                                             # Printing an empty line
# -----Printing Merged Lists---------

# Create an instance of the Solution class
solution = Solution()

# Call the mergeTwoLists method with the sample linked lists
merged_list = solution.mergeTwoLists(l1, l2)

print("Merged List:")

# Print the merged list                                                 # JFR : 
while merged_list and merged_list.next:                                 # 1st iteration - while merged_list and merged_list.next:
    print(merged_list.val, end=" -> ")                                  # 2nd iteration - while merged_list.next and merged_list.next.next:
    merged_list = merged_list.next                                      # 3rd iteration - while merged_list.next.next and merged_list.next.next.next:
print(merged_list.val)  # Print the last element of the merged list     # and so on......  You can check by - print(merged_list.next.val) - add .next accordingly
```

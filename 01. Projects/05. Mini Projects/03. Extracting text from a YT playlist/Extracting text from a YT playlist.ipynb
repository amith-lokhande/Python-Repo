````python
# Specify the URL of the webpage
url = 'https://www.youtube.com/playlist?list=PLot-Xpze53lfQmTEztbgdp8ALEoydvnRQ'

# Configure the Selenium webdriver
# Here we use 'webdriver' to load and extract from the webpage as evident in the below steps
driver = webdriver.Chrome()

# Load the webpage
driver.get(url)

# Extract the page source
page_source = driver.page_source

# Parse the HTML source using BeautifulSoup
soup = BeautifulSoup(page_source, 'html.parser')

# Extract all the text from the webpage
all_text = soup.get_text()

# Print the extracted text
#print(all_text)          # Commenting this print as the o/p is very large

# Quit the webdriver.     # This is done in order to free up the resources
driver.quit()
````

````python
text = all_text

# Remove leading and trailing spaces
text = text.strip()

# Split the text into lines
lines = text.splitlines()

# Remove empty lines
lines = [line for line in lines if line.strip() != ""]

# Join the lines back into a single string
result = "\n".join(lines)

#print(result)  # Commenting this print as the o/p is large
````

#### <span style="background-color: yellow;">Explaining : lines = [line for line in lines if line.strip() != ""]</span>


* **`line.strip()`** removes any leading or trailing whitespace characters from a line.
* **`line.strip() != ""`** checks if the stripped line is not an empty string.
* The list comprehension **`[line for line in lines if line.strip() != ""]`** iterates over each line in the lines list and filters out any lines that are empty (contain only whitespace characters). It creates a new list with the non-empty lines.

````python
# Displaying the first 50 lines of the o/p (optional)

text = all_text.strip()
lines = text.splitlines()
lines = [line for line in lines if line.strip() != ""]

# Print the first three lines
print('\n'.join(lines[:50]))
````

````python
# Example 2 - Modified code to load the entire webpage

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time

# Specify the URL of the webpage
url = 'https://www.youtube.com/playlist?list=PLot-Xpze53lfOdF3KwpMSFEyfE77zIwiP'

# Configure the Selenium webdriver
driver = webdriver.Chrome()

# Load the webpage
driver.get(url)

# Scroll down until reaching the bottom of the page
while True:
    # Get the current page height
    current_height = driver.execute_script("return document.documentElement.scrollHeight")
    
    # Scroll to the bottom of the page
    driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
    
    # Wait for a short interval to allow content to load
    time.sleep(2)                                              # We can chage the number accordingly. The given time is in secs.
    
    # Calculate the new page height after scrolling
    new_height = driver.execute_script("return document.documentElement.scrollHeight")
    
    # Check if the page height has not changed after scrolling
    if new_height == current_height:
        break

# Extract the page source
page_source = driver.page_source

# Parse the HTML source using BeautifulSoup or perform further operations
soup = BeautifulSoup(page_source, 'html.parser')

# Extract all the text from the webpage
all_text = soup.get_text()

# Print the extracted text
#print(all_text)               # Commenting this print as the o/p is very large

# Quit the webdriver
driver.quit()                  # This is done in order to free up the resources

# ---

text = all_text

# Remove leading and trailing spaces
text = text.strip()

# Split the text into lines
lines = text.splitlines()

# Remove empty lines
lines = [line for line in lines if line.strip() != ""]

# Join the lines back into a single string
result = "\n".join(lines)

#print(result)  # Commenting this print as the o/p is large

# ---

text = result

lines = text.splitlines()
now_playing_lines = []

for i, line in enumerate(lines):
    if re.search(r"Now playing", line):
        now_playing_lines.append(lines[i + 1])

for line in now_playing_lines:
    print(line.strip())
````

#### <span style="background-color: yellow;">Explaining : </span>
#### <span style="background-color: yellow;">1. current_height = driver.execute_script - ("return document.documentElement.scrollHeight") </span>

In the line `document.documentElement.scrollHeight`, `document` refers to the DOM (Document Object Model) of the webpage. It represents the HTML document structure and provides a way to interact with the elements of the webpage.

`documentElement` refers to the root element of the document, which is typically the `<html>` tag. The `scrollHeight` property of `documentElement` gives the total height of the webpage, including any overflow content that may be hidden and requires scrolling to be viewed.

So, `document.documentElement.scrollHeight` retrieves the current height of the entire webpage, allowing us to determine the scrolling position and whether we have reached the bottom of the page.

#### <span style="background-color: yellow;">2. driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")</span>

The code `driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")` is using the Selenium WebDriver's `execute_script()` method to execute a JavaScript code snippet in the browser.

In this case, the JavaScript code `window.scrollTo(0, document.documentElement.scrollHeight)` is being executed. Let's break it down:

- `window.scrollTo()` is a JavaScript function used to scroll the browser window to a specified position.
- The first parameter `0` indicates the horizontal scrolling position. In this case, it is set to `0` to scroll to the leftmost position.
- The second parameter `document.documentElement.scrollHeight` represents the vertical scrolling position. `document.documentElement` refers to the root element of the HTML document (usually the `<html>` tag), and `scrollHeight` is a property that gives the total height of the document, including any content that may require scrolling to be viewed.

Therefore, the code is instructing the browser to scroll to the bottom of the webpage by setting the vertical scrolling position to the total height of the document.

#### <span style="background-color: yellow;">3. html tag </span>

The `<html>` tag is an element in HTML (Hypertext Markup Language) that represents the root of an HTML document. It is the top-level element and encloses all other elements in the document.

The `<html>` tag serves as the container for all the content of a webpage. It provides the structure and defines the document type. Inside the `<html>` tag, you will typically find two main sections:

1. `<head>`: This section contains metadata about the document, such as the document title, character encoding, linked stylesheets, JavaScript files, and other information that is not directly visible on the webpage.

2. `<body>`: This section contains the visible content of the webpage, including text, images, links, headings, paragraphs, and other HTML elements that are rendered in the browser.

Here's an example of a basic HTML document structure with the `<html>`, `<head>`, and `<body>` tags:

```html
<!DOCTYPE html>
<html>
<head>
  <title>My Webpage</title>
  <!-- Additional metadata and linked files -->
</head>
<body>
  <!-- Visible content of the webpage -->
  <h1>Welcome to My Webpage</h1>
  <p>This is the content of my webpage.</p>
  <!-- Other HTML elements -->
</body>
</html>
```

In summary, the `<html>` tag represents the root element of an HTML document and encapsulates the entire structure and content of the webpage.

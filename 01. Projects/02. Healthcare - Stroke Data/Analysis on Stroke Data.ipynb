**Additional Notes:**

**a. Correlation Strength:** 

The strength of the correlation can be interpreted as follows:

`r = 0.1 to 0.3:` Weak correlation. The variables have a weak and possibly non-linear relationship.

**b. Causality:**

There are two conditions for causality i.e.,

`1. Significant Correlation -` There must be a significant correlation between the 2 variables.

`2. One of the below 2 conditions must be satisfied -` 

`a. Temporal precedence -` The cause must precede the effect in time. This means that changes in the cause variable should occur before changes in the effect variable. For example, let's consider the variables BMI and AGL (avg. glucose level), there is no precedence as to which variable's data should be collected first. 

`b. Profound Theory -` There must be a theoretically founded and plausible theory which indicates in which direction the causal relationship goes.

**Calculating p-values for Hypothesis Tests**

```python
# Calculation Part ------------------------------------------------------

# Load the df required for correlation -> df_corr 
df_corr = df[['age', 'avg_glucose_level', 'bmi']].copy()

# Rename the columns
df_corr.rename(columns={'age': 'Age', 'avg_glucose_level': 'AGL', 'bmi': 'BMI'}, inplace=True)

# Compute the correlation matrix using pandas corr method
correlation_matrix = df_corr.corr(method='spearman')

# Calculate p-values for the correlation coefficients using scipy spearmanr function
p_values = df_corr.corr(method=lambda x, y: stats.spearmanr(x, y)[1])

# Figure Part -----------------------------------------------------------

# Create a 2x2 grid for subplots using Gridspec Layout
fig = plt.figure(figsize=(12, 6))
gs = fig.add_gridspec(1, 2, width_ratios=[1, 1.25], height_ratios=[1])

# Adding subplots to the GridSpec object
ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])

# Create the heatmap for the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='Set3', cbar=False, ax=ax1)
ax1.set_title('Correlation Matrix')

# Create the heatmap for the p-values
sns.heatmap(p_values, annot=True, cmap='Set3', ax=ax2)
ax2.set_title('p-values Matrix')

# Show the plots
plt.tight_layout()
plt.show()
```

**Calculating p-values for Hypothesis Tests**

```python
# We conduct chi-square tests for all the Hypotheses as all the variables involved are categorical and indpendent in nature

# Hypothesis 4 - 

# Variables for contingency table
a = len(df[(df['hypertension'] == 1) & (df['stroke'] == 1)])
b = len(df[(df['hypertension'] == 1) & (df['stroke'] == 0)])
c = len(df[(df['hypertension'] == 0) & (df['stroke'] == 1)])
d = len(df[(df['hypertension'] == 0) & (df['stroke'] == 0)])

# Create the contingency table
contingency_table = [[a, b], [c, d]]

# Perform the Chi-square test
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

# Print the results
print("Chi-square Test    : Hypothesis 4\n")
print("Test Statistic     :", format(chi2, ".2f")) # Chi-square statistic
print("Degrees of Freedom :", dof)  # DOF = (r-1)(c-1) where, r-rows, c-columns 
print("P-value            :", format(p_value, ".2e")) # Comparing chi-square statistc and dof in chi-square chart distribution to get the p-values.
print("Expected Frequencies :\n", expected,"\n") # all expected frequencies must be greater than 5 to fulfill chi-square assumptions 

# ----------------------------------------------------------------------------

# Hypothesis 5 -

# Variables for contingency table
a = len(df[(df['heart_disease'] == 1) & (df['stroke'] == 1)])
b = len(df[(df['heart_disease'] == 1) & (df['stroke'] == 0)])
c = len(df[(df['heart_disease'] == 0) & (df['stroke'] == 1)])
d = len(df[(df['heart_disease'] == 0) & (df['stroke'] == 0)])

# Create the contingency table
contingency_table = [[a, b], [c, d]]

# Perform the Chi-square test
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

# Print the results
print("Chi-square Test    : Hypothesis 5\n")
print("Test Statistic     :", format(chi2, ".2f"))
print("Degrees of Freedom :", dof)
print("P-value            :", format(p_value, ".2e")) 
print("Expected Frequencies :\n", expected,"\n")

# ----------------------------------------------------------------------------

# Hypothesis 6 -

# Create the contingency table
contingency_table = pd.crosstab(df['stroke'], df['smoking_status'])

# Perform the Chi-square test
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

# Print the results
print("Chi-square Test      : Hypothesis 6\n")
print("Test Statistic       :", format(chi2, ".2f"))
print("Degrees of Freedom   :", dof)
print("P-value              :", format(p_value, ".2e")) 
print("Expected Frequencies :\n", expected) 
```

`Hypothesis 5 - Heart Disease and Stroke`
